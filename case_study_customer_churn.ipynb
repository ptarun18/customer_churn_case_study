```python
# first we have th find "q1" and "q3" values

q1=df["seniorcitizen"].quantile(0.25)
print("q1:",q1)

q3=df["seniorcitizen"].quantile(0.75)
print("q3:",q3)
```

```python
# then we have to find "IQR" range
# IQR= q3-q1

iqr=q3-q1
print("iqr:",round(iqr,2))
```

```python
# then we have to find lowest and highest values

lowest= q1-1.5*iqr
print("lowest:",lowest)

highest= q3+1.5*iqr
print("highest:",round(highest,1))
```
